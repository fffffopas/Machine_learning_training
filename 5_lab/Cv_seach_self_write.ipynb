```python
import numpy as np
from xgboost import XGBRegressor
import joblib
```

```python
X_train, X_test, y_train, y_test, X, y = joblib.load("../Data/preprocessing_all_info.joblib")
```

```python
from Self_realisation.GridSearchCV_realisation import GridSearchCV
from Self_realisation.RandomSearchCV_realisation import RandomSearchCV
from Self_realisation.BayesSearchCV_realisation import BayesSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import loguniform, randint
from skopt.space import Real, Integer
```

```python
params_grid = {"n_estimators" : [250 * i for i in range(1, 4)], "max_depth" : [3, 5, 6], "learning_rate" : [0.1, 0.01]}
params_space_random = {"n_estimators" : randint(250, 1000), "max_depth" : randint(3, 10), "learning_rate" : loguniform(1e-3, 1e-1)}
params_space_bayes = {"n_estimators" : randint(250, 1000), "max_depth" : Integer(3, 10), "learning_rate" : loguniform(1e-3, 1e-1)}
```

```python
base_model = XGBRegressor(subsample=0.7, random_state=101)
```

```python
grid_model = GridSearchCV(
    estimator=base_model,
    params_grid=params_grid,
    scoring='r2',
    random_state=101
)

grid_model.fit(X_train, y_train)
```

```python
y_pred = grid_model.predict(X_test)

MSE = mean_squared_error(y_pred, y_test)
MAE = mean_absolute_error(y_pred, y_test)
RMSE = np.sqrt(MSE)
R2 = r2_score(y_true=y_test, y_pred=y_pred)

print(f"MAE:{MAE}, MSE:{MSE}, RMSE:{RMSE}, R^2:{R2}")
```

```python
random_model = RandomSearchCV(
    estimator=base_model,
    params_space=params_space_random,
    n_iter=15, 
    scoring='r2',
    random_state=101,
)

random_model.fit(X_train, y_train)
```

```python
y_pred = random_model.predict(X_test)

MSE = mean_squared_error(y_pred, y_test)
MAE = mean_absolute_error(y_pred, y_test)
RMSE = np.sqrt(MSE)
R2 = r2_score(y_true=y_test, y_pred=y_pred)

print(f"MAE:{MAE}, MSE:{MSE}, RMSE:{RMSE}, R^2:{R2}")
```

```python
bayes_model = BayesSearchCV(
    estimator=base_model,
    params_space=params_space_bayes,
    n_iter=15,
    scoring='r2',
    random_state=101
)

bayes_model.fit(X_train, y_train)
```

```python
y_pred = bayes_model.predict(X_test)

MSE = mean_squared_error(y_pred, y_test)
MAE = mean_absolute_error(y_pred, y_test)
RMSE = np.sqrt(MSE)
R2 = r2_score(y_true=y_test, y_pred=y_pred)

print(f"MAE:{MAE}, MSE:{MSE}, RMSE:{RMSE}, R^2:{R2}")
```
