```python
import numpy as np
```

```python
class Linear_regression():
    def __init__(self):
        self.weight = None
    
    def fit(self, X_train, y_train):
        b_col = np.ones((X_train.shape[0], 1))
        X_train = np.concat((X_train, b_col), axis=1)

        left_part = X_train.T @ X_train
        rigth_part = X_train.T @ y_train
        w = np.linalg.solve(left_part,rigth_part)

        self.weight = w

    def predict(self, X_test):
        b_col = np.ones((X_test.shape[0], 1))
        X_test = np.concat((X_test, b_col), axis=1)
        return X_test @ self.weight
```

```python
def test_linear_exact_fit():
    model = Linear_regression()
    X = np.array([[1], [2], [3], [4]])
    y = np.array([3, 5, 7, 9])
    
    model.fit(X, y)
    
    expected_weights = np.array([2.0, 1.0])
    np.testing.assert_allclose(model.weight, expected_weights, atol=1e-10)
    
    X_new = np.array([[5]])
    prediction = model.predict(X_new)
    assert np.isclose(prediction[0], 11.0)

test_linear_exact_fit()
```

```python
def test_multidimensional_shapes():
    model = Linear_regression()
    N, D = 100, 5  
    X = np.random.randn(N, D)
    true_w = np.random.randn(D + 1) 
    
    X_with_b = np.hstack([X, np.ones((N, 1))])
    y = X_with_b @ true_w
    
    model.fit(X, y)
    
    assert len(model.weight) == D + 1
    
    X_test = np.random.randn(10, D)
    preds = model.predict(X_test)
    assert preds.shape == (10,)
```
