```python
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
```

```python
X_train, X_test, y_train, y_test, X, y = joblib.load("../Data/preprocessing_all_info.joblib")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

### Дерево 1


```python
model_1 = RandomForestRegressor(n_estimators=200, max_depth=10, min_samples_leaf=10, max_features='sqrt', n_jobs=-1, random_state=101)
```

```python
model_1.fit(X_train, y_train)
```

```python
y_pred = model_1.predict(X_test)
MSE = mean_squared_error(y_pred, y_test)
MAE = mean_absolute_error(y_pred, y_test)
RMSE = np.sqrt(MSE)
R = r2_score(y_test, y_pred)

print(f"MAE: {MAE}, MSE: {MSE}, RMSE: {RMSE}, R: {R}")
```

```python
residuals = y_test - y_pred
plt.hist(residuals, bins=50)
plt.title('Распределение ошибок')
plt.show()
```

```python
print(model_1.predict(X) ,y[0])
```
