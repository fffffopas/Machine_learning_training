```python
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("../Data/2_lab.csv", index_col=False)
```

```python
df.head()
```

```python
df.columns
```

```python
model = SentenceTransformer("all-MiniLM-l6-v2")
```

```python
embendings = model.encode(df["name"].astype(str).tolist(), batch_size=64, show_progress_bar=True)
```

```python
embendings.shape
```

```python
n_comp = 32
pca = PCA(n_components=n_comp)
pca_features = pca.fit_transform(embendings)
```

```python
pca_features.shape
```

```python
pca_cols = [f'name_pca_{i}' for i in range(n_comp)]
df_pca = pd.DataFrame(pca_features, columns=pca_cols, index=df.index)

df_ = pd.concat([df, df_pca], axis=1)
```

```python
len(df_.columns)
```

```python
df_.to_csv("../Data/2_lab_proc.csv", index=False)
```
